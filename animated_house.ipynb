# Animation with matrix transformations

The function `animated_house()` displays an animation obtained by applying a sequence of matrix transformations to an image of a house. Each frame of the animation shows the image after applying to it some matrix transformation. Matrices defining these transformations change only a little from one movie frame to the next one, which creates an illusion of a smooth movement. 

The variable `delay` controls the speed of the animation: the bigger the value, the longer each frame will be displayed. 

```python
ani = animated_house(delay = 100)
```
