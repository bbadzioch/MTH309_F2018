# Color mixing

```python
import matplotlib.pyplot as plt
import numpy as np

def color_mixer(img, A):
    
    A = np.array(A).astype(float)
    
    if img.dtype == 'uint8':
        img = img.astype(float)/255
    
    new_img = np.transpose(np.dot(A, np.transpose(img, axes = (0, 2, 1))), axes = (1, 2, 0))
    new_img[new_img < 0] = 0
    new_img[new_img > 1] = 1

    return (new_img*255).astype('uint8')


def color_sample(rgb):
    if np.min(rgb) < 0 or np.max(rgb) > 255:
        print("ERROR: RGB coordinates must be intergers in the range 0-255.")
        return None
    rgb = np.array(rgb).astype('uint8')
    sample = np.ones((10,10, 3), dtype='uint8')*rgb
    plt.figure(figsize = (4,4))
    ax = plt.gca()
    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_title("R {}    G {}    B {}".format(rgb[0], rgb[1], rgb[2]))
    ax.imshow(sample)
    plt.show()
```

```python
x = plt.imread("py309/magritte.jpg")
```

```python
plt.figure(figsize=(8,5))
plt.imshow(x)
plt.show()
```

## Adjusting white balance

```python
A = diag(255/223, 255/210, 255/219)
plt.figure(figsize=(8,5))
y = color_mix(x, A)
plt.imshow(y)
plt.show()
```
