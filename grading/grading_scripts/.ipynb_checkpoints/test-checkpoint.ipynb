## Project 7: GPS

## Introduction

The purpose of this report is to representing the process of a Global Positioning System which pinpoints a precise location of an unknown target relative to the distances of eight satellites and their positions. This process is dependent on two key functions. The first is a gps function that generates a value of a randomly passed point in relation to the eight satellites. This value is a representation of how far away the randomly selected point is from the unknown point being located. The second is a function which produces a set number of randomly generated points and generates the value from the gps function based on each of those points. This function begins with some arbitrary starting point, then maps to a new point that is closer to the unknown location. This process contains a couple of variables which can be adjusted in order to produce a final result more effectively. Examining the different impact of these variables being altered is a primary goal in this report, in addition to the reproduction of the overall system.

### Required User-Defined Functions

The following are the two functions needed to model the GPS triangulation process for pinpointing an unknown location.

`gps(x, y)` takes in an x and y position values to then be compared to each of the eight satellites to produce a value representing the accuracy of the passed in point being at the same location as the unknown point. This is accomplished be accumulating a sum of values produced using each of the satellites which is the difference of the distance from the passed in point to the satellite and the distance of the unknown point and the satellite. The distance from the passed in point to the satellite is determined using the Pythagorean Theorem, then all values are squared to eliminate negative values.

`stumbledown(f, x, y, stp, nstp)` is a function which handles collecting the steps needed to locate the unknown point. This is accomplished by starting at an arbitrary location, such as (0, 0), then obtaining the gps value. This value represents some distance that the tested location is from the unknown location. Then a series of random points are generated, tested, and counted as the next step in line on the progression to reaching the unknown location provided the gps value generated is less than the gps value of the previous step, and therefore closer to the unknown. Each new random point is generated in any direction on the two-dimensional space, though with a range limited to a `step size` of the previous successful step. The `step size` is initially set to some value, then will shrink in half if there are more than 20 consecutive fails. All details of each attempt is collected, including if it was successful or not, which are all then passed back to the caller when complete. Complication is determined at a minimum by a maximum step count passed as the final argument, however, at any point when the step size is less than 0.0002, it is determined that the accuracy of the current position is sufficient and the process will terminate.

<p style="font-style:italic; color:red;">
More precisely, <code style="color:red; font-weight:bold;">stumbledown</code> is a general purpose function searching for minimum of a function <code style="color:red; font-weight:bold;">f</code> which is passed to 
<code style="color:red; font-weight:bold;">stumbledown</code> as one of the arguments. It is not explicitly related to the <code style="color:red; font-weight:bold;">gps</code> function. 
</p>

```python
def gps(x, y):
    dsum = 0
    for i in range(0, 8):
        ds = ((((x - sat[0][i])**2) + ((y - sat[1][i])**2)) ** 0.5)
        ds = (ds - sat[2][i]) ** 2
        dsum += ds
    return dsum
```

<p style="font-style:italic; color:red;">
The code above works, but a good practice is to use in a function only data which is either defined in the body of the function, or passed to the function as an argument. The satellite data should be passed to the <code style="color:red; font-weight:bold;">gps</code> function in one of these two ways. 
</p>

```python
def stumbledown(f, x, y, stp, nstp):
    ds = f(x, y)  # Initial GPS value
    p = [x, y]    # x, y paired up
    q = [[],[]]   # Position to be tested and compared to last
    sd = []       # Final list of all positions tried, with corresponding gps value, success on taking step, and step size used
    fails = 0     # Setup counter for check on number of fails

    sd.append([x, y, ds, 1, stp]) # First value details
    for i in range(nstp):
        if ds == 0: # Bail if gps value is perfectly on point to unknown value
            break
        bol = 0 # Default setting on success/fail marker
        d = (np.random.rand(2) - 0.5) * 2 * stp # New random point between -1 and 1 * current step size
        q = p - d # New point in relation to last successful step
        dc = f(q[0], q[1]) # GPS value of random point
        if dc < ds: # If new GPS value is better than GPS value of last successful step, step to new position
            fails = 0 # Reset fail counter
            bol = 1   # Set success/fail marker to success
            ds = dc   # Replace GPS value with new
            p = q     # Set new last successful to current random
        fails += 1 # Increment fail counter
        if fails > 20: # Dynamically adjust step size if too many fails
            stp = stp * 0.5 # Reduces step size by half if consecutive fails exceeds 20
        
        sd.append([p[0], p[1], ds, bol, stp]) # Collect details of attempt
        if stp < 0.0002: # Step sizes below this level indicate accuracy of pinpointing unknown location is sufficient and process may terminate early
            break
    return sd
```

<p style="font-style:italic; color:red;">
The code <code style="color:red; font-weight:bold;">q = [[],[]]</code> is not needed. 
</p>

<p style="font-style:italic; color:red;">
The condition <code style="color:red; font-weight:bold;">if ds == 0: break</code> is fine for the <code style="color:red; font-weight:bold;">gps</code> function, but will create problems if <code style="color:red; font-weight:bold;">stumbledown</code> is used to search for minimum of other functions that may have negative values. 
</p>

<p style="font-style:italic; color:red;">
It is prefereable to use <code style="color:red; font-weight:bold;">True</code> and <code style="color:red; font-weight:bold;">False</code> rather than 1 and 0 for boolean values. It makes the code more readable. 
</p>

<p style="font-style:italic; color:red;">
It would be better to make the minimal step size one of the arguments of the <code style="color:red; font-weight:bold;">stumbledown</code> function instead of setting it to 0.0002 in the code of the function: in this way this value could be more easily changed if greater/lesser precision is required. 
</p>


## Plot Analysis

### Step Size of 1000

This plot is the first test of using a particular step size to produce some outcome and analyze the results. As an overview, the plot is made primarily of the satellita data and gps function. The eight red dots represent the locations of the satellites, which the red rings surrounding them are the distances from each satellite and the unknown target. It can be seen that at the point close to (400, 400), all eight rights meet, indicating the unknown location to be at this position. The gps function is represented by the underlying contour plot, which coincides with the details of the satellites, where the darker blue area is closest to the unknown point and red represents areas farthers from the unknown point. 

Focusing now on the random steps leading to the target, each step is represented by a white dot, with the corresponding step size indicated with a white square. It can be seen easily this way that by using a large step size of 1000, the random position checks cover a large portion of the available space, allowing for good odds in a given random point to not only be closer to the unknown target than a previous step, it may also be a point that is much closer to the target than an attempt made with a smaller step size. However, such a large step size is only useful in the beginning attempts, and as each new successful step nears closer to the unknown target, the significant amount of available space to attempt a following point becomes a greater hinderence than a help. This is due to the limited number of points in the giant step size which would be closer to the target than the previous step. Because of this then, it is necessary for the step size to be reduced in size, in this case, by cutting the size in half. It appears this reduction in size was necessary a few times before the next successful step could be located as the next size down from the giant squares is closer to an eigth or a sixtenth rather than a half.

In the end, with a step size of 1000, only 314 of the 10000 steps were needed to be used, and of those tries, 56 were successful. This produced an efficiency calculation of 17.83%.

```python
%matplotlib inline

# X, Y, Z setup
x = np.linspace(-200, 1000, 1000)
y = np.linspace(-200, 1000, 1000)
X, Y = np.meshgrid(x, y)
Z = gps(X, Y)
cnts = [0, 0]
stps = [1000, 10000] # Step Size / Max number of steps to try

# Plot setup
plt.figure(figsize=(10, 8))
ax = plt.axes()
plt.xlim(-200,1000)
plt.ylim(-200,1000)

# Satellite locations - circles representing distance to unknown target
for i in range(0, 8):
    plt.plot(sat[0][i], sat[1][i], 'ro')
    ax.add_patch(patches.Circle((sat[0][i], sat[1][i]), radius = sat[2][i], fc='none', ec='r',  linewidth = 1))

# Random mapping from given point tracing to unknown location
sd = stumbledown(gps, 0, 0, stps[0], stps[1])
for i in range(len(sd)):
    if sd[i][3] == 1:
        cnts[1] += 1 # Count of successful steps
        plt.plot(sd[i][0], sd[i][1], 'w.')
        ax.add_patch(patches.Rectangle((sd[i][0]-(sd[i][4]/2), sd[i][1]-(sd[i][4]/2)), width=sd[i][4], height=sd[i][4], fc='none', ec='w',  linewidth = 1))

cnts[0] = len(sd) # Count of steps taken
eff = format(cnts[1]/cnts[0], '.2%') # Efficiency - steps taken / steps checked
# Contour plot based on gps function with satellites
CS = plt.contourf(X, Y, Z, 60, cmap=cm.jet, antialiased=True)
plt.title('Step Size = {}  |  Steps Checked = {}   Steps Taken = {}  |  EFF = {}'.format(stps[0], cnts[0], cnts[1], eff), fontsize=10)
plt.contour(X, Y, Z, 60, colors='w', linewidths=0.4, antialiased=True)
plt.colorbar(CS)
plt.show()
```

<p style="font-style:italic; color:red;">
This is a nice plot, but it should also include information what were the coordinates of the user location computed by the <code style="color:red; font-weight:bold;">stumbledown</code> function.

## Conclusion

Based on the results of these varied step sizes, it appears the best size is at or near a value of 50. An additional observation is that due to the dynamic sizing used in the stumbledown function, it is better to attempt a larger step size than one that is smaller. However, there is a perfect range, such as the case of 50, where a balance can be found of adequate space to produce random steps that are succesful, while not landing too far away from the direction that heads towards the unknown target. It is also apparent that the number 50 is specific to this particular set of data and that had the values of the satellite positions and distances been 10 or 1000 times greater, it is expected that results closer to the 10 or 1 sizes shown here would have been produced.
