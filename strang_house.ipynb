# Strang house

Strang house is a program illustrating the effect of linear transformations. Drag the red and blue dots to deform the drawing of a house according to a linear transformation. The matrix of the resulting transformation is  displayed in the upper left corner. The semi-transparent drawing is the image of the house before the transformation. 

**Note:** The command `%run py309/strangehouse` below can be replaced by one of 

- `%run py309/strangehouse symmetric`
- `%run py309/strangehouse diagonal`
- `%run py309/strangehouse orthogonal`

This will limit linear transformations to these represented by, respectively, symmetric, diagonal or orthogonal matrices.
